```python
def categorize(x, bins, labels):
    for i in range(0, len(bins)):
        if x < bins[i]: return i
    return len(bins)
    
sLength = len(df['pregnant'])
df['pregnant-norm'] = df['pregnant'].apply(lambda x: categorize(x, [2, 6], ['low', 'medium', 'high']))
df['plasma-glucose-norm'] = df['plasma-glucose'].apply(lambda x: categorize(x, [90, 151], ['low', 'medium', 'high']))
df['diastolic-bp-norm'] = df['diastolic-bp'].apply(lambda x: categorize(x, [80, 91], ['normal', 'normal-to-high', 'high']))
df['bmi-norm'] = df['bmi'].apply(lambda x: categorize(x, [25, 31, 36], ['low', 'normal', 'obese', 'severly-obese']))
df['dpf-norm'] = df['dpf'].apply(lambda x: categorize(x, [0.4, 0.81], ['low', 'medium', 'high']))
df['age-norm'] = df['age'].apply(lambda x: categorize(x, [40, 60], ['below-40', '40-plus', '60-plus']))

df
```

```python
from math import sqrt
import scipy.stats      
dist = scipy.stats.norm(0, 1)

def h_test(count, avg, std):
    sp = ((count[0]-1) * std[0]**2 + (count[1]-1) * std[1]**2)/(count[0] + count[1]-2)
    z = abs(avg[0]-avg[1])/(sqrt(sp * (1.0/count[0] + 1.0/count[1])))
    p = dist.cdf(z)

    if p > 0.5:
        p = 1-p
        
    return (z, p)
```
